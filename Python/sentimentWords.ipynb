```python
data = pd.read_csv("HindiSentiWordnet.txt", delimiter=' ')
fields = ['POS_TAG', 'ID', 'POS', 'NEG', 'LIST_OF_WORDS']
```

```python
data
```

```python
words_dict = {}
for i in data.index:
    words = data[fields[4]][i].split(',')
    for word in words:
          words_dict[word] = (data[fields[2]][i], data[fields[3]][i])
```

```python
words_dict
```

```python
pos=[]
neg=[]
words_list=[]
```

```python
for word,tup in words_dict.items() :
      p, n = tup
      sum = p - n
      if sum > 0:
        pos.append(1)
        neg.append(0)
        words_list.append(word)
      elif sum < 0:
        pos.append(0)
        neg.append(1)
        words_list.append(word)
      else:
        pos.append(0)
        neg.append(0)
        words_list.append(word)
```

```python
dict = {'LIST_OF_WORDS': words_list, 'POS': pos, 'NEG': neg}

df = pd.DataFrame(dict)
```
